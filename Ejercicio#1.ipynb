```python
import pandas as pd

df = pd.read_csv('./ventas.csv')
#Ver cantidad de filas del dataframe
df.shape
df.head()
```

```python
Cantida_categoria = df.groupby('Producto')['Cantidad'].sum()
Cantida_categoria
```

```python
Mayor_ventas = Cantida_categoria.sort_values(ascending=False).head(1)
Mayor_ventas
```
